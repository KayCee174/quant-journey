Exercise 1.1 - Create a Matrix: Generate a 10 \times 3 NumPy array of random floats representing price changes for 3 different assets over 10 days.

```python
price = np.round(np.random.rand(10,3),2)
price
```

Exercise 1.2 - Normalization: Subtract the mean of each column from itself (centering the data).

```python
# mean for each column 

column1 = price[:,0] 
mean_column1 = np.mean(column1)
print(f'The Mean of column1: {mean_column1}')

column2 = price[:,1]
mean_column2 = np.mean(column2)
print(f'The Mean of column2: {mean_column2}')

column3 = price[:,2]
mean_column3 = np.mean(column3)
print(f'The Mean of column3: {mean_column3}')
```

```python
# Subtracting the mean of each column from itself 

price_change1 = column1 - mean_column1
print(f'The price change for column1: {price_change1}')

price_change2 = column2 - mean_column2
print(f'The price change for column2: {price_change2}')

price_change3 = column3 - mean_column3
print(f'The price change for column3: {price_change3}')
```

```python
# how to arrange it back to a 10,3

prices1 = np.array(price_change1)
prices2 = np.array(price_change2)
prices3 = np.array(price_change3)

new_price = np.array([[prices1],[prices2],[prices3]])
new_price
```

Exercise 1.3 - Risk Calc: Calculate the Standard Deviation for each asset (column) to see which is the most volatile.

```python
# Standard Deviation for each asset(column)

sd_column1 = np.std(column1)
print(f'The Standard Deviation for column1: {sd_column1}')

sd_column2 = np.std(column2)
print(f'The Standard Deviation for column2: {sd_column2}')

sd_column3 = np.std(column3)
print(f'The Standard Deviation for column3: {sd_column3}')

max_sd = [sd_column1,sd_column2,sd_column3]
print(f'The most volatile: {np.max(max_sd)}')
```
