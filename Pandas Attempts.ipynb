# The Quant Challenge (Bonus)

```python
import numpy.linalg as la
```

```python
import numpy as np
import numpy.linalg as la

# Step 1: choose assets
assets = ['EURUSD', 'GBPUSD']
df2 = df[df['Asset'].isin(assets)]

# Step 2: extract Profit_USD as NumPy arrays
eur = df2[df2['Asset'] == 'EURUSD']['Profit_USD'].to_numpy()
gbp = df2[df2['Asset'] == 'GBPUSD']['Profit_USD'].to_numpy()

# Step 3: align lengths
min_len = min(len(eur), len(gbp))
eur = eur[:min_len]
gbp = gbp[:min_len]

# Step 4: standardize
eur_z = (eur - eur.mean()) / eur.std()
gbp_z = (gbp - gbp.mean()) / gbp.std()

# Step 5: correlation using linalg-style approach
corr = (eur_z * gbp_z).mean()

print(f'the correlation between EURUSD & GBPUSD is {corr:.4f}')
```
