```python
import numpy as np
import pandas as pd
```

Exercise 1 - Create a Matrix: Generate a 10 \times 3 NumPy array of random floats representing price changes for 3 different assets over 10 days.

```python
# 3 assets over 10 days - 3 columns, 2 rows

price = np.round(np.random.rand(10,3),4)
price
```

Exercise 2 - Normalization: Subtract the mean of each column from itself (centering the data).

```python
mean_column1 = np.mean(price[:,0])
mean_column2 = np.mean(price[:,1])
mean_column3 = np.mean(price[:,2])


new_price = price - [mean_column1, mean_column2, mean_column3]
np.round(new_price,4)
```

Exercise 3 - Risk Calc: Calculate the Standard Deviation for each asset (column) to see which is the most volatile.

```python
std_column1 = np.round(np.std(price[:,0]),4)
std_column2 = np.round(np.std(price[:,1]),4)
std_column3 = np.round(np.std(price[:,1]),4)

max_std = np.max([std_column1,std_column2,std_column3])

# this isn't necessry suggest a shorter means
if max_std == std_column1:
    print(f'the most volatile column is column1, at {max_std}')
elif max_std == std_column2:
    print(f'the most volatile column is column2, at {max_std}')
elif max_std == std_column3:
    print(f'the most volatile column is column3, at {max_std}')
```
